```
tokenizer = MosesTokenizer()
N = 5
```

```
data_df = pd.read_csv('/content/drive/MyDrive/demetre_{pipia, uridia}/data/paragraph_all_only_georgian_shuffled.csv', lineterminator='\n', nrows = utils.NROWS)
data_df.shape, data_df.head()
```

```
train_df, validate_df, test_df = np.split(data_df.sample(frac=1, random_state=42), [int(.70*len(data_df)), int(.85*len(data_df))])
```

```
def n_gram_train_model(model, paragraphs, tokenizer):
    iteration = 0
    for paragraph in paragraphs:
        model.update(tokenizer(paragraph))

        if iteration % 20000 == 0:
            print('Iteration: ' + str(iteration))
        
        iteration += 1
```

```
model = utils.NgramModel(N)
```

```
n_gram_train_model(model, train_df['Paragraph'].tolist(), tokenizer)
```
